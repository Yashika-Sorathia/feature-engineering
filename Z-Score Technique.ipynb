```python
# detecting and removing outliers using z-score
# z-score -> shows how many std away a data point is
import pandas as pd

df = pd.read_csv('heights1.csv')
df.head()
```

```python
# create a column for z-score
df['zscore'] = (df.height - df.height.mean())/df.height.std()
df.head()
```

```python
# detecting outlier
# z-score > 3 are outliers same as upper limit
df[df.zscore > 3]
```

```python
# z-score < -3 are outliers same as lower limit
df[df.zscore < -3]
```

```python
df[(df.zscore > 3) | (df.zscore < -3)]
```

```python
# removing outliers
df1 = df[(df.zscore < 3)&(df.zscore > -3)]
df1.head()
```
